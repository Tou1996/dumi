```python
# our next task is is isin()
# find the most destructive batsman in the ipl history
# we find bases of strike rate (number of runs / number of balls *100)
# min batsman faced 200 ball in 16 - 20 over
```

```python
death_over=delivery[delivery['over']>15]
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
# runs scored by these 43 batsman
# balls faced by these batsman

final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls =final.groupby('batsman')['batsman_runs'].count()
```

```python
Strike_rate=((runs/balls)*100).sort_values(ascending=False)
Strike_rate
```
